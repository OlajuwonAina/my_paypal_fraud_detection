## Machine Learning

### Data Preprocessing and Model Traning

To ensure better performance, I plan to use the Robust Scaler to scale down the data.


```python
# splitting the data before scaling it 
X = data.drop(['Class'], axis = 1)
y = data['Class']
```

```python
#Applying Standard Scaler

scaler = RobustScaler()
X_scaled = scaler.fit_transform(X)
X_scaled.shape
```

```python
# defining the classifier for test the sampling methods
clf = LogisticRegression()
```

```python
def method_validation(X_data, y_data, classifier):
    # Splitting the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, random_state=42, stratify=y_data)

    # Train the classifier on the provided data
    classifier.fit(X_data, y_data)

    # Use predict_proba to obtain probability predictions for the test data
    y_predicted_probabilities = classifier.predict_proba(X_test)[:, 1]

    # Calculate precision, recall, and thresholds using precision_recall_curve
    precision_values, recall_values, threshold_values = precision_recall_curve(y_test, y_predicted_probabilities)

    # Calculate the AUPRC score using the auc function
    auprc_score = auc(recall_values, precision_values)

    # Print the AUPRC score, mean precision, and mean recall
    print('AUPRC: %.3f' % auprc_score)
    print("Mean Precision:", precision_values.mean())
    print("Mean Recall:", recall_values.mean())

    # Plot the precision-recall curve
    plt.plot(recall_values, precision_values, label='Precision-Recall Curve')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.legend()
    plt.show()
```

```python
# Checking classifier performance
method_validation(X_scaled, y, clf)
```

The AUPRC score, indicating a value of 0.725, implies suboptimal performance in detecting fraudulent transactions. The observed deficiency can be linked to the imbalanced distribution within the dataset, emphasizing the imperative to rectify this imbalance for improved classifier accuracy.

### Handling imbalance in the dataset 

Will be using 2 methods.
- Oversampling method
- Smote method

#### Oversampling Method

```python
# Initialize the RandomOverSampler object
ros = RandomOverSampler(random_state=42)

# Resample the data
X_resampled, y_resampled = ros.fit_resample(X_scaled, y)

# Print the number of samples in each class
print("Class 0 samples:", sum(y_resampled==0))
print("Class 1 samples:", sum(y_resampled==1))

# checking the AUPRC score for this method
method_validation(X_resampled, y_resampled, clf)
```

A performance score of 0.989 in AUPRC suggests a strong ability to accurately detect instances of the fraudulent class. This implies the model exhibits high precision and recall in identifying the majority of fraudulent transactions.

#### Smote Method

```python
# Define the SMOTE model
smote = SMOTE()

# Fit the SMOTE model to the data
X_resampled, y_resampled = smote.fit_resample(X_scaled, y)

# Print the number of samples in each class
print("Class 0 samples:", sum(y_resampled==0))
print("Class 1 samples:", sum(y_resampled==1))

# checking the AUPRC score for this sampling method method
method_validation(X_resampled, y_resampled, clf)
```

An AUPRC score of 0.991 signifies strong performance in identifying fraudulent transactions, highlighting the model's ability to attain elevated precision and recall values. This implies the efficacy of the <b>SMOTE</b> technique in addressing dataset imbalance and enhancing the classifier's accuracy for fraudulent transaction detection.
